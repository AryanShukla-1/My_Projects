## Implement Launch Site Dropdown and Pie Chart Callback

### Subtask:
Add the dropdown input component for launch sites. Implement a callback function to render a success-pie-chart based on the selected launch site from the dropdown. This pie chart will show the success rate of launches for the chosen site.


**Reasoning**:
I will import the necessary Dash dependencies and Plotly Express, then define a callback function that takes the selected launch site from the dropdown as input and generates a pie chart showing the launch success rate, addressing the current subtask instructions.



```python
import plotly.express as px
from dash.dependencies import Input, Output

# Add the callback function
@app.callback(
    Output(component_id='success-pie-chart', component_property='figure'),
    Input(component_id='launch-site-dropdown', component_property='value')
)
def get_pie_chart(entered_site):
    filtered_df = spacex_df
    if entered_site == 'All Sites':
        fig = px.pie(filtered_df,
                     values='Launch outcome',
                     names='Launch site',
                     title='Total Success Launches By Site')
    else:
        # Filter data for the selected site
        filtered_df = spacex_df[spacex_df['Launch site'] == entered_site]
        # Count success and failure for the selected site
        outcome_counts = filtered_df['Launch outcome'].value_counts().reset_index()
        outcome_counts.columns = ['Launch outcome', 'count']

        fig = px.pie(outcome_counts,
                     values='count',
                     names='Launch outcome',
                     title=f'Total Success and Failure Launches for {entered_site}')
    return fig

print("Pie chart callback defined successfully.")
```

## Implement Payload Range Slider and Scatter Plot Callback

### Subtask:
Add the range slider component to select the payload range. Implement a callback function to render the success-payload-scatter-chart. This scatter plot will visualize the relationship between payload, launch success, and potentially other variables, filtered by the selected payload range and launch site.


**Reasoning**:
I will define a Dash callback function that takes the selected launch site and payload range as input to generate a scatter plot visualizing the relationship between payload, flight number, launch outcome, and booster landing status, addressing the current subtask instructions.



```python
import plotly.express as px
from dash.dependencies import Input, Output

@app.callback(
    Output(component_id='success-payload-scatter-chart', component_property='figure'),
    [Input(component_id='launch-site-dropdown', component_property='value'),
     Input(component_id='payload-slider', component_property='value')])
def get_scatter_chart(entered_site, payload_range):
    low, high = payload_range

    # Filter by payload range first
    filtered_payload_df = spacex_df[ (spacex_df['Payload mass'] >= low) & (spacex_df['Payload mass'] <= high) ]

    if entered_site == 'All Sites':
        fig = px.scatter(filtered_payload_df,
                         x='Payload mass',
                         y='Flight No.',
                         color='Booster landing',
                         symbol='Launch outcome',
                         title=f'Payload vs. Launch Outcome for All Sites (Payload: {low}kg - {high}kg)')
    else:
        # Filter by launch site if not 'All Sites'
        filtered_site_payload_df = filtered_payload_df[filtered_payload_df['Launch site'] == entered_site]
        fig = px.scatter(filtered_site_payload_df,
                         x='Payload mass',
                         y='Flight No.',
                         color='Booster landing',
                         symbol='Launch outcome',
                         title=f'Payload vs. Launch Outcome for {entered_site} (Payload: {low}kg - {high}kg)')
    return fig

print("Scatter plot callback defined successfully.")
```

```python
import pandas as pd
import dash
from dash import dcc
from dash import html
import plotly.express as px
from dash.dependencies import Input, Output
from google.colab.output import eval_js
from google.colab import output

# Load the SpaceX data
spacex_df = pd.read_csv('/content/spacex_web_scraped.csv')

# Data Preprocessing for Payload mass
spacex_df['Payload mass'] = spacex_df['Payload mass'].str.replace(' kg', '', regex=False).str.replace(',', '', regex=False)
spacex_df['Payload mass'] = pd.to_numeric(spacex_df['Payload mass'], errors='coerce')

# Get unique launch sites for the dropdown
launch_sites = spacex_df['Launch site'].unique().tolist()
launch_site_options = [{'label': 'All Sites', 'value': 'All Sites'}]
for site in launch_sites:
    launch_site_options.append({'label': site, 'value': site})

# Get min and max payload mass for the range slider from valid data
valid_payload_mass = spacex_df['Payload mass'].dropna()
min_payload = int(valid_payload_mass.min()) if not valid_payload_mass.empty else 0
max_payload = int(valid_payload_mass.max()) if not valid_payload_mass.empty else 10000

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the app layout
app.layout = html.Div(children=[
    html.H1('SpaceX Launch Records Dashboard', style={'textAlign': 'center', 'color': '#503D36', 'font-size': 40}),

    # Dropdown for launch sites
    html.Div([
        html.Label("Select Launch Site:", style={'marginRight': '10px'}),
        dcc.Dropdown(
            id='launch-site-dropdown',
            options=launch_site_options,
            value='All Sites',
            placeholder="Select a Launch Site here",
            searchable=True
        )
    ]),
    html.Br(),

    # Div for pie chart
    html.Div(dcc.Graph(id='success-pie-chart')),
    html.Br(),

    # Range slider for payload mass
    html.Div([
        html.Label("Payload Mass (kg):"),
        dcc.RangeSlider(
            id='payload-slider',
            min=0,
            max=10000,
            step=1000,
            marks={i: f'{i} kg' for i in range(0, 10001, 1000)},
            value=[min_payload, max_payload]
        )
    ]),
    html.Br(),

    # Div for scatter plot
    html.Div(dcc.Graph(id='success-payload-scatter-chart'))
])

# Pie chart callback
@app.callback(
    Output(component_id='success-pie-chart', component_property='figure'),
    Input(component_id='launch-site-dropdown', component_property='value')
)
def get_pie_chart(entered_site):
    filtered_df = spacex_df
    if entered_site == 'All Sites':
        # For 'All Sites', calculate success rate for all launches
        success_counts = filtered_df[filtered_df['Launch outcome'] == 'Success'].groupby('Launch site').size().reset_index(name='count')
        total_counts = filtered_df.groupby('Launch site').size().reset_index(name='total')
        merged_counts = pd.merge(total_counts, success_counts, on='Launch site', how='left').fillna(0)
        merged_counts['Success_Rate'] = (merged_counts['count'] / merged_counts['total']) * 100

        fig = px.pie(merged_counts,
                     values='Success_Rate',
                     names='Launch site',
                     title='Total Success Launches By Site')
    else:
        # Filter data for the selected site
        filtered_df = spacex_df[spacex_df['Launch site'] == entered_site]
        # Count success and failure for the selected site
        outcome_counts = filtered_df['Launch outcome'].value_counts().reset_index()
        outcome_counts.columns = ['Launch outcome', 'count']

        fig = px.pie(outcome_counts,
                     values='count',
                     names='Launch outcome',
                     title=f'Total Success and Failure Launches for {entered_site}')
    return fig

# Scatter plot callback
@app.callback(
    Output(component_id='success-payload-scatter-chart', component_property='figure'),
    [Input(component_id='launch-site-dropdown', component_property='value'),
     Input(component_id='payload-slider', component_property='value')])
def get_scatter_chart(entered_site, payload_range):
    low, high = payload_range

    # Filter by payload range first
    filtered_payload_df = spacex_df[ (spacex_df['Payload mass'] >= low) & (spacex_df['Payload mass'] <= high) ]

    if entered_site == 'All Sites':
        fig = px.scatter(filtered_payload_df,
                         x='Payload mass',
                         y='Flight No.',
                         color='Booster landing',
                         symbol='Launch outcome',
                         title=f'Payload vs. Launch Outcome for All Sites (Payload: {low}kg - {high}kg)')
    else:
        # Filter by launch site if not 'All Sites'
        filtered_site_payload_df = filtered_payload_df[filtered_payload_df['Launch site'] == entered_site]
        fig = px.scatter(filtered_site_payload_df,
                         x='Payload mass',
                         y='Flight No.',
                         color='Booster landing',
                         symbol='Launch outcome',
                         title=f'Payload vs. Launch Outcome for {entered_site} (Payload: {low}kg - {high}kg)')
    return fig

# Run the app
if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0', port=8050)

```
