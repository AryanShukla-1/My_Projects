## Task 1: Display Unique Launch Sites

### Subtask:
Display the names of the unique launch sites in the space mission.


**Reasoning**:
The subtask is to display the unique launch sites. This can be achieved by executing a SQL query to select distinct values from the 'Launch_Site' column in the 'SPACEXTABLE'.



```python
%%sql
SELECT DISTINCT "Launch_Site" FROM SPACEXTABLE;
```

## Task 2: Display Launch Sites starting with 'CCA'

### Subtask:
Display 5 records where launch sites begin with the string 'CCA' by executing a SQL query to select 5 records from 'SPACEXTABLE' where 'Launch_Site' LIKE 'CCA%'.


**Reasoning**:
The subtask requires displaying 5 records where the launch site starts with 'CCA'. This can be achieved by writing a SQL query to select all columns from 'SPACEXTABLE', filtering for 'Launch_Site' like 'CCA%', and limiting the results to 5.



```python
%%sql
SELECT * FROM SPACEXTABLE
WHERE "Launch_Site" LIKE 'CCA%'
LIMIT 5;
```

**Reasoning**:
The subtask is to display the total payload mass carried by boosters launched by NASA (CRS). This can be achieved by summing the 'Payload_Mass__kg_' column from the 'SPACEXTABLE' where the 'Customer' is 'NASA (CRS)'.



```python
%%sql
SELECT SUM("Payload_Mass__kg_") FROM SPACEXTABLE WHERE Customer = 'NASA (CRS)';
```

## Average Payload Mass for F9 v1.1

### Subtask:
Display the average payload mass carried by booster version F9 v1.1.


**Reasoning**:
The subtask is to display the average payload mass carried by booster version F9 v1.1. This can be achieved by executing a SQL query to calculate the average of the 'Payload_Mass__kg_' column from the 'SPACEXTABLE' where the 'Booster_Version' is 'F9 v1.1'.



```python
%%sql
SELECT AVG("Payload_Mass__kg_") FROM SPACEXTABLE WHERE "Booster_Version" = 'F9 v1.1';
```

## Task 5: Date of First Successful Ground Pad Landing

### Subtask:
List the date when the first successful landing outcome in 'ground pad' was achieved.


**Reasoning**:
The subtask is to find the date of the first successful landing outcome on a 'ground pad'. This can be achieved by executing a SQL query to select the minimum 'Date' from the 'SPACEXTABLE' where the 'Landing_Outcome' is 'Success (ground pad)'.



```python
%%sql
SELECT MIN(Date) FROM SPACEXTABLE WHERE "Landing_Outcome" = 'Success (ground pad)';
```

## Boosters with Successful Drone Ship Landing and Specific Payload Mass

### Subtask:
List the names of the boosters which have a 'Success (drone ship)' landing outcome and a payload mass greater than 4000 but less than 6000.


**Reasoning**:
The subtask requires listing booster versions based on specific landing outcomes and payload mass ranges. A SQL query is needed to select distinct booster versions from the 'SPACEXTABLE', filtering by 'Landing_Outcome' being 'Success (drone ship)' and 'Payload_Mass__kg_' being between 4000 and 6000.



```python
%%sql
SELECT DISTINCT "Booster_Version"
FROM SPACEXTABLE
WHERE "Landing_Outcome" = 'Success (drone ship)' AND "Payload_Mass__kg_" > 4000 AND "Payload_Mass__kg_" < 6000;
```

## Total Successful and Failure Mission Outcomes

### Subtask:
List the total number of successful and failure mission outcomes.


**Reasoning**:
To list the total number of successful and failure mission outcomes, I need to count the occurrences of each unique mission outcome from the 'SPACEXTABLE' and group them by 'Mission_Outcome'.



```python
%%sql
SELECT "Mission_Outcome", COUNT("Mission_Outcome") AS Outcome_Count
FROM SPACEXTABLE
GROUP BY "Mission_Outcome";
```

## Task 8: Booster Versions with Maximum Payload Mass

### Subtask:
List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


**Reasoning**:
The subtask is to list booster versions that carried the maximum payload mass. This can be achieved by using a subquery to first find the maximum payload mass, and then selecting the booster versions from the main table that match this maximum payload mass.



```python
%%sql
SELECT DISTINCT "Booster_Version"
FROM SPACEXTABLE
WHERE "Payload_Mass__kg_" = (
    SELECT MAX("Payload_Mass__kg_")
    FROM SPACEXTABLE
);
```

**Reasoning**:
The next subtask is to list records showing month names, failure landing outcomes in drone ship, booster versions, and launch sites for the year 2015. This requires filtering by date and landing outcome, and extracting the month from the date. As SQLite does not support `MONTHNAME`, `substr(Date, 6,2)` will be used to extract the month, and `substr(Date,0,5)='2015'` to filter by year.



```python
%%sql
SELECT
  substr(Date, 6, 2) AS Month,
  "Landing_Outcome",
  "Booster_Version",
  "Launch_Site"
FROM SPACEXTABLE
WHERE
  substr(Date, 0, 5) = '2015' AND "Landing_Outcome" LIKE 'Failure (drone ship)%';
```

## Task 10: Rank Landing Outcomes by Count

### Subtask:
Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.

**Reasoning**:
To rank the landing outcomes by count within the specified date range, I need to execute a SQL query that filters records by date, groups them by 'Landing_Outcome', counts the occurrences, and then orders these counts in descending order.



```python
%%sql
SELECT "Landing_Outcome", COUNT("Landing_Outcome") AS Outcome_Count
FROM SPACEXTABLE
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY "Landing_Outcome"
ORDER BY Outcome_Count DESC;
```
